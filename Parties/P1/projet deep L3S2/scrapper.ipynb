```python
import random
import requests
import json
def get_useragent():
    user_agent_list = [
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36",
        "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.141 Safari/537.36",
        "Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/535.1 (KHTML, like Gecko) Chrome/13.0.782.112 Safari/535.1",
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.3"]
    
    hasard=random.choice(user_agent_list)
    return(hasard)
```

```python
def get_page(url):
	response = requests.get(url, headers={"User-Agent": get_useragent()})
	html = response.text
	return html
```

```python
def get_hyperlien(competition):
    html = get_page(competition)
    if html == []:
        return []
    split1 = html.split('data-url=')
    splits = []
    for i in split1 :
        split2 = i.split('title="Download"')
        split2 = str(split2)
        split2 = split2.split('"http://www.w3.org/2000/svg" viewBox="0 0 21.596 24.363" class="w-4">\\n')[0]
        split2 = split2.split("<svg xmlns=")[0]
        split2 = split2.split("', '")[0]
        split2 = split2.lstrip("['")
        split2 = split2.replace('"','')
        split2 = split2.replace("Overlay",'')
        if split2.startswith('https://'):
            splits.append(split2)
        
    return splits
```

```python
import librosa
import os

def mp3_to_spec(path):
    try: 
        audio, sr = librosa.load(f"{path}", sr=22050)
        stft = librosa.stft(audio)
        spectrogram = np.abs(stft)
        log_spectrogram = librosa.amplitude_to_db(spectrogram)
    except Exception as e:
        print(f"[attention] Erreur sur {path} : {e}")
        return None, None
    return log_spectrogram

def couper_les_fichiers_mp3_a_1_minute(log_spectogram):
    
    nouveau_log_spectogram = log_spectogram[:, :2000] # 2000 frames pour environ 1 minute
    return nouveau_log_spectogram

def fichier_mp3_to_fichier_spec_avec_supprimage(path_mp3, path_spec): # ca marche 
    for fichier in os.listdir(path_mp3):
        i=0
        noms = ["Malt.Tabulated Sounds - Chill.mp3" , "Open Work Stocking - Chill"]
        if fichier.endswith(".mp3"):  # Vérifier si le fichier est un fichier .mp3
            fichier_mp3 = os.path.join(path_mp3, fichier)
            """print(os.path.getsize(fichier_mp3))"""
            if os.path.getsize(fichier_mp3) > 10000 and fichier not in noms :    
                i+=1
                fichier_spec = os.path.join(path_spec, f"{os.path.splitext(fichier)[0]}.npy")
                # Convertir le fichier mp3 en spectrogramme
                log_spectrogram = mp3_to_spec(fichier_mp3)
                if log_spectrogram[0] is None:
                    continue
                log_spectrogram = couper_les_fichiers_mp3_a_1_minute(log_spectrogram)
                # Sauvegarder le spectrogramme en tant que fichier .npy
                np.save(fichier_spec, log_spectrogram)
                """print(f"Spectrogramme sauvegardé : {fichier_spec}")"""
                """fichier_mp3_to_fichier_spec("C:\\Users\\gabri\\Desktop\\dauphine\\L3\\S2\\Deep_learning\\projet deep L3S2\\musique", "C:\\Users\\gabri\\Desktop\\dauphine\\L3\\S2\\Deep_learning\\projet deep L3S2\\spec")         
                """


    # Parcourir tous les fichiers dans le dossier
    for fichier in os.listdir(path_mp3):
        file_path = os.path.join(path_mp3, fichier)
        # Vérifier si c'est un fichier avant de le supprimer
        if os.path.isfile(file_path):
            os.remove(file_path)
            print(f"Supprimé : {file_path}")
```
